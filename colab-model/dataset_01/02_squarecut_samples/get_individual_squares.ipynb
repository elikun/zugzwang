```
# Configure matplotlib
%matplotlib inline
from matplotlib import pyplot as plt

import cv2
import os

# Utility class
# Generalized image display function
def display_image(img):
  if img.ndim == 2:
    print("Displaying Grayscale image:")
    plt.imshow(img, cmap='gray')
  else:
    print("Displaying Color image:")
    plt.imshow(img)
  plt.show()

# Take sample data No. xx
sample_num = "20"
raw_sample_name = "sample_" + sample_num + ".jpg"
raw_sample_dir = "/gdrive/My Drive/Colab Notebooks/project-checkmate/01_raw_sample_data/"
preprocessed_sample_dir = "/gdrive/My Drive/Colab Notebooks/project-checkmate/02_preprocessed_sample_data/" + "sample_" + str(sample_num) + "/"
try: 
    os.mkdir(preprocessed_sample_dir)
except OSError as error:
    print(error)

raw_sample_filename = raw_sample_dir + raw_sample_name
cv_img = cv2.imread(raw_sample_filename)
cv_img_400x400 = cv_img[40:440, 120:520]
display_image(cv_img_400x400)
print(cv_img_400x400[0:50, 0:50].shape)
display_image(cv_img_400x400[0:50, 0:50])
for i in range(0, 351, 50):
  for j in range(0, 351, 50):
    print('cv_img_400x400[' + str(i) + ':' + str(i+50) + ', ' + str(j) + ':' + str(j+50) + ']')
    cv_save = cv_img_400x400[i:i+50, j:j+50, :]
    cv2.imwrite(preprocessed_sample_dir + str(sample_num) + '-' + str(round(i/50)) + '-' + str(round(j/50)) + '.jpg', cv_save)
```
