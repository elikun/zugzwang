```
# Configure matplotlib
%matplotlib inline
from matplotlib import pyplot as plt

import cv2

# Utility class
# Generalized image display function
def display_image(img):
  if img.ndim == 2:
    print("Displaying Grayscale image:")
    plt.imshow(img, cmap='gray')
  else:
    print("Displaying Color image:")
    plt.imshow(img)
  plt.show()

cv_img = cv2.imread('/gdrive/My Drive/Colab Notebooks/project-checkmate/train15.jpg')
#cv_img = cv2.rectangle(cv_img, (120, 40), (520, 440), (255, 0, 0), 1)
cv_img_400x400 = cv_img[40:440, 120:520]
#print(cv_img[40:440, 120:520].shape)
display_image(cv_img_400x400)
print(cv_img_400x400[0:50, 0:50].shape)
display_image(cv_img_400x400[0:50, 0:50])
#cv2.imwrite('/gdrive/My Drive/Colab Notebooks/project-checkmate/train_data/train01/01.jpg', cv_img_400x400[0:50, 0:50, :])
for i in range(0, 351, 50):
  for j in range(0, 351, 50):
    print('cv_img_400x400[' + str(i) + ':' + str(i+50) + ', ' + str(j) + ':' + str(j+50) + ']')
    cv_save = cv_img_400x400[i:i+50, j:j+50, :]
    #cv2.imwrite('/gdrive/My Drive/Colab Notebooks/project-checkmate/train_data/train01/01.jpg', cv_save)
    cv2.imwrite('/gdrive/My Drive/Colab Notebooks/project-checkmate/train15/15-' + str(round(i/50)) + '-' + str(round(j/50)) + '.jpg', cv_save)
#display_image(cv_img[41:440, 121:520])
```
